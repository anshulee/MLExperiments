```python
#return files from given path and all its subdirectories to the lowest level
def ReturnDataFiles(path):
    print("checking path :-",path)
    filesInPath=[]
    for f in os.listdir(path):
        if os.path.isdir(os.path.join(path,f)):
           dataFiles=ReturnDataFiles(os.path.join(path,f))
           for file in dataFiles:
                filesInPath.append(file)
        else :
            if(os.path.isfile(os.path.join(path,f))):
                filesInPath.append(os.path.join(path,f))
    print("files found at ", path, ":-",filesInPath)
    return filesInPath
```

```python
#read files from path and load
def ReadMergeAndLoadFileFromPath(path='datasets\List'):
    print (os.listdir('datasets\List'))
    dirPath='datasets\List'
    files = [f for f in os.listdir(dirPath) if os.path.isfile(os.path.join(dirPath,f))]
    print(files)
    merged = []

    for f in files:
        filePath=os.path.join(dirPath,f)
        filename, ext = os.path.splitext(filePath)
        print('ext')
        if ext == '.csv':
            print('found csv')
            try:
                read = pd.read_csv(filePath,error_bad_lines=False,sep=';')
                merged.append(read)
            except:
                print('error for :- ',filePath)
    result = pd.concat(merged)
   
    result.to_csv(os.path.join("MergedOutput",'merged.csv'))
```

```python
#find all columns for "points in" and "Sales in"
pointsIn= csvFile.columns[csvFile.columns.str.contains(pat = 'Points in')] 


salesIn= csvFile.columns[csvFile.columns.str.contains(pat = 'Sales in')]
print(pointsIn,salesIn)
```

```python
#process all the salesIn and PointsIn columns and make them ready for addingup
for point in pointsIn:
    csvFile[point].fillna(0, inplace=True)
for sale in salesIn:
    csvFile[sale].fillna(0,inplace=True)
csvFile.head()
```

```python
def CalculatePoints(row):
    try:
        global pointsIn
        result= 0.0
        for point in pointsIn:
            result=float(result)+float(row[point])
        
    except:
        result=0
    return result
    
    
```

```python
def CalculateSales(row):
    try:
        global salesIn
        result= 0.0
        for point in salesIn:
            result=float(result)+float(row[point])
    except:
        result=0
    return result
```

```python
csvFile['Points'] = csvFile.apply(lambda row: CalculatePoints(row), axis=1)
```

```python
csvFile['Sales'] = csvFile.apply(lambda row: CalculateSales(row), axis=1)
```

```python
#Convert Zip and store Zip to string
csvFile['Zip']= csvFile['Zip'].astype(str)
csvFile['Store Zip']= csvFile['Store Zip'].astype(str)
```

```python
def distanceFromStore(row):
    try:
        global invalidZipCount
        
        storeZip= row['Store Zip']


        userZip= row['Zip']
        
        userZipObject= zipcode.isequal(userZip)
      
    
        storeZipObject= zipcode.isequal(storeZip)

        

        storeAddress=(storeZipObject.lat,storeZipObject.lon)

        userAddress= (userZipObject.lat,userZipObject.lon)

        distance= haversine.haversine(storeAddress,userAddress)
        
       
    except:
        print('Exception detected. ZipCode was :- ',row['Zip'])
        distance=-1
        invalidZipCount=invalidZipCount+1
        
    return distance
    
   
   
  
```

```python

csvFile['Distance']=csvFile.apply (lambda row: distanceFromStore (row),axis=1)
print (invalidZipCount)
```
